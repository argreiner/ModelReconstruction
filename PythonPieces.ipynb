```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.optimize as optim
```

```python
# Recursive programming
def fibonacci(n):
    if n == 0: return 0
    if n == 1: return 1
    return fibonacci(n-1)+fibonacci(n-2)
# And now the factorial
```

```python
# Root finding
def fun(x,*argv):
    a=argv
    return x - a*np.cos(x)

sol1 = optim.root(fun, x0 = 0,args=(2.))
sol2 = optim.root(fun, x0 = -2.,args=(3.))
print("root of f(x,2.): x =", sol1.x)
print("root of f(x,3.): x =", sol2.x)
```

```python
%matplotlib notebook
x=np.linspace(-3.,3.)
plt.plot(x,fun(x,2.))
plt.plot(x,fun(x,3.))
plt.plot(np.linspace(-3.5,3.5,10),np.zeros(10))
```

```python
%matplotlib notebook
#Get the libraries
import numpy as np
import matplotlib.pyplot as plt
#
f=np.zeros(100) # Define the solution vector, 
f[45:55]=1.0 # set initial conditions ,
# choose Delta t
dt=.4 # that's a good choice
# dt = 0.6 # this starts to oscillate #dt=1. #thisblowsup
# The above must be at the first row or the cell
timesteps =1000 # number 
for i in np.arange(timesteps):
    fl=np.roll(f,1) 
    fr=np.roll(f,-1) 
    fc=f.copy() 
    f=f+(fl-2*fc+fr)*dt
#of timesteps
# loop
# shift right f
# shift left f
# copy in order not to overwrite # caluate one timestep forward
    if i%100 == 0: 
        plt.plot(f) # plot every 100. timestep 
        plt.savefig('myDifusion'+str(i)+'.png')
plt.show() # show the plots
```
